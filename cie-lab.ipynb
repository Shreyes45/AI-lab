```python
def solve_jug_puzzle_dfs():
    capacities = (12, 8, 5)
    target = 6
    initial_state = (12, 0, 0)
    visited = set()
    stack = [(initial_state, [initial_state])]
    while stack:
        current_state, path = stack.pop()
        if target in current_state:
            return path
        if current_state in visited:
            continue
        visited.add(current_state)
        j1, j2, j3 = current_state
        c1, c2, c3 = capacities
        next_states = []
        next_states.append((c1, j2, j3) if j1 < c1 else (j1, j2, j3))
        next_states.append((j1, c2, j3) if j2 < c2 else (j1, j2, j3))
        next_states.append((j1, j2, c3) if j3 < c3 else (j1, j2, j3))
        next_states.append((0, j2, j3) if j1 > 0 else (j1, j2, j3))
        next_states.append((j1, 0, j3) if j2 > 0 else (j1, j2, j3))
        next_states.append((j1, j2, 0) if j3 > 0 else (j1, j2, j3))
        pour_amount = min(j1, c2 - j2)
        next_states.append((j1 - pour_amount, j2 + pour_amount, j3))
        pour_amount = min(j1, c3 - j3)
        next_states.append((j1 - pour_amount, j2, j3 + pour_amount))
        pour_amount = min(j2, c1 - j1)
        next_states.append((j1 + pour_amount, j2 - pour_amount, j3))
        pour_amount = min(j2, c3 - j3)
        next_states.append((j1, j2 - pour_amount, j3 + pour_amount))
        pour_amount = min(j3, c1 - j1)
        next_states.append((j1 + pour_amount, j2, j3 - pour_amount))
        pour_amount = min(j3, c2 - j2)
        next_states.append((j1, j2 + pour_amount, j3 - pour_amount))
        for state in next_states:
            if state not in visited:
                new_path = list(path)
                new_path.append(state)
                stack.append((state, new_path))
    return None

result = solve_jug_puzzle_dfs()
if result:
    for i, state in enumerate(result):
        print(f"Step {i}: {state}")
else:
    print("No solution found.")
```
